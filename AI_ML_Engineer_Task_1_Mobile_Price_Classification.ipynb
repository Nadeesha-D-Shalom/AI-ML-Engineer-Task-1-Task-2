```python
import pandas as pd

# Load dataset
file_path = "Mobile_Price_Classification.csv"
data = pd.read_csv(file_path)
```

```python
# Display basic information
print("Dataset Shape:", data.shape)
print("\nColumns:")
print(data.columns)
```

```python
# Separate features and target
X = data.drop("price_range", axis=1)
y = data["price_range"]
```

```python
print("\nFeatures Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
# Preview first few rows
print("\nFirst 5 rows of features:")
print(X.head())
```

```python
print("\nFirst 5 target values:")
print(y.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
# Verify shapes
print("Training Features Shape:", X_train.shape)
print("Training Labels Shape:", y_train.shape)

print("Testing Features Shape:", X_test.shape)
print("Testing Labels Shape:", y_test.shape)
```

# Import tensorflow

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

#Build the model

```python
# Build ANN model
model = Sequential()

# First hidden layer (8 neurons)
model.add(Dense(units=8, activation="relu", input_dim=20))

# Second hidden layer (4 neurons)
model.add(Dense(4, activation="relu"))

# Output layer (binary classification)
model.add(Dense(1, activation="sigmoid"))

# Model summary
model.summary()
```

# Compile the model

```python
model.compile(
    optimizer="sgd",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

print("Model compiled successfully.")
```

# Train the model

```python
history = model.fit(
    X_train,
    y_train,
    epochs=100,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)
```

```python
# Save model weights
weights_path = "mobile_price_ann.weights.h5"
model.save_weights(weights_path)

print(f"Model weights saved successfully at: {weights_path}")
```

Load weights

```python
model.load_weights("mobile_price_ann.weights.h5")
```
